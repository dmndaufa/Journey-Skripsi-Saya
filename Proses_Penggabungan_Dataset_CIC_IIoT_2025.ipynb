```python
import os
import tarfile
import pandas as pd

# 1. Pastikan path ini benar sesuai lokasi folder Anda di Google Drive
base_path = '/content/drive/My Drive/Dataset/CIC_IIoT_2025/'

# List kosong untuk menampung semua dataframe yang sudah digabung per time window
all_dfs = []

print("Memulai proses ekstraksi dan penggabungan berurutan...")

# 2. Loop dari 1 sampai 10 (untuk 1sec hingga 10sec)
for i in range(1, 11):
    print(f"\n--- Memproses Time Window: {i} detik ---")

    try:
        # 3. Tentukan nama file attack dan benign untuk iterasi saat ini
        attack_filename = f'attack_samples_{i}sec.csv.tar.xz'
        benign_filename = f'benign_samples_{i}sec.csv.tar.xz'

        attack_filepath = os.path.join(base_path, 'attack_data', attack_filename)
        benign_filepath = os.path.join(base_path, 'benign_data', benign_filename)

        # List untuk data attack dan benign pada time window ini
        current_window_dfs = []

        # --- Proses File Attack ---
        print(f"  - Mengekstrak dan membaca {attack_filename}...")
        with tarfile.open(attack_filepath, 'r:xz') as tar:
            csv_name = tar.getnames()[0]
            tar.extractall(path='/content/temp_csv/')
            df_attack = pd.read_csv(os.path.join('/content/temp_csv/', csv_name))
            df_attack['time_window'] = i # Tambahkan kolom time_window
            current_window_dfs.append(df_attack)

        # --- Proses File Benign ---
        print(f"  - Mengekstrak dan membaca {benign_filename}...")
        with tarfile.open(benign_filepath, 'r:xz') as tar:
            csv_name = tar.getnames()[0]
            tar.extractall(path='/content/temp_csv/')
            df_benign = pd.read_csv(os.path.join('/content/temp_csv/', csv_name))
            df_benign['time_window'] = i # Tambahkan kolom time_window
            current_window_dfs.append(df_benign)

        # Gabungkan data attack dan benign untuk time window saat ini
        combined_df_for_window = pd.concat(current_window_dfs, ignore_index=True)
        all_dfs.append(combined_df_for_window)

        print(f"  - Selesai menggabungkan data {i} detik. Total baris: {len(combined_df_for_window)}")

    except FileNotFoundError as e:
        print(f"    WARNING: File tidak ditemukan, melewati time window {i} detik. Detail: {e}")
    except Exception as e:
        print(f"    ERROR: Terjadi kesalahan pada time window {i} detik. Detail: {e}")

# 5. Gabungkan semua hasil dari setiap time window menjadi satu DataFrame final
print("\n--- Menggabungkan semua hasil dari semua time window ---")
if all_dfs:
    df_final = pd.concat(all_dfs, ignore_index=True)
    print("\nProses selesai!")
    print("Jumlah total baris dan kolom gabungan:", df_final.shape)
    print("Distribusi data berdasarkan time_window:")
    print(df_final['time_window'].value_counts().sort_index())
    df_final.head()
else:
    print("Tidak ada data yang berhasil diproses.")
```

```python
# Tentukan path dan nama file tujuan di Google Drive
save_path_parquet = '/content/drive/My Drive/Dataset/CIC_IIoT_2025/final_dataset.parquet'

print(f"Menyimpan DataFrame sebagai Parquet ke: {save_path_parquet}")
df_final.to_parquet(save_path_parquet)

print("Selesai menyimpan sebagai Parquet!")
```
