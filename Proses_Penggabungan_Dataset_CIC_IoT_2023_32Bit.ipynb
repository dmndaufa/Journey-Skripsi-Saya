## Langkah 3:  Fungsi untuk Downcasting

```python
def optimize_types(df):
    """
    Fungsi untuk downcast tipe data (menghemat RAM).
    Secara spesifik mengubah kolom int yang mungkin dibaca sebagai float
    (karena NaN) ke tipe Int32 (nullable).
    """

    # 1. Tentukan kolom mana yang HARUS menjadi Nullable Integer (Int32)
    nullable_int_cols = [
        'Protocol Type', # Dari info asli
        'ack_count',
        'syn_count',
        'fin_count',
        'rst_count',
        'Tot sum',
        'Min',
        'Max',
        'Number'
    ]

    # 2. Tentukan kolom mana yang harus menjadi float32
    float64_cols = df.select_dtypes(include=['float64']).columns

    # 3. Konversi kolom Nullable Integer
    for col in nullable_int_cols:
        if col in df.columns:
            try:
                df[col] = df[col].astype('Int32')
            except Exception as e:
                print(f"Peringatan: Gagal konversi {col} ke Int32: {e}")

    # 4. Konversi kolom Float lainnya
    for col in float64_cols:
        if col not in nullable_int_cols:
            try:
                df[col] = df[col].astype('float32')
            except Exception as e:
                print(f"Peringatan: Gagal konversi {col} ke float32: {e}")

    # 5. Konversi int64 lain (jika ada)
    remaining_int64_cols = df.select_dtypes(include=['int64']).columns
    for col in remaining_int64_cols:
        if col not in nullable_int_cols:
             try:
                df[col] = df[col].astype('Int32')
             except Exception as e:
                print(f"Peringatan: Gagal konversi {col} (sisa) ke Int32: {e}")

    return df
```

## Langkah 4: Proses Inti (Load, Label, Concat)

```python
print("--- Memulai Proses Penggabungan Dataset (100% Data, 32-bit) ---")
final_ddf_list = []

for class_name, glob_pattern in path_map.items():

    files = glob.glob(glob_pattern)
    if not files:
        print(f"Peringatan: Tidak ada file ditemukan untuk pola '{glob_pattern}' (Kelas: {class_name}). Dilewati.")
        continue

    print(f"Memproses {class_name} (menemukan {len(files)} file, sampling=100%) ...")

    # Muat data ke Dask (tetap dengan perbaikan dtype float64)
    ddf = dd.read_csv(files, dtype={
        'http.request.method': 'object',
        'http.user_agent': 'object',
        'Number': 'float64',
        'Max': 'float64',
        'Min': 'float64',
        'Tot sum': 'float64',
        'ack_count': 'float64',
        'fin_count': 'float64',
        'rst_count': 'float64',
        'syn_count': 'float64',
        })

    # Terapkan downcasting ke 32-bit
    ddf = ddf.map_partitions(optimize_types)

    # !!! TAMBAHKAN KOLOM LABEL !!!
    new_label_name = label_rename_map.get(class_name)

    if new_label_name is None:
        print(f"Peringatan: Nama label baru untuk '{class_name}' tidak ditemukan di label_rename_map!")
        new_label_name = class_name

    ddf['Label'] = new_label_name

    final_ddf_list.append(ddf)
```

## Langkah 5: Menggabungkan dan Menyimpan ke Tipe Parquet

```python
if not final_ddf_list:
    print("KESALAHAN: Tidak ada data yang diproses. Harap periksa 'path_map' Anda.")
else:
    print("--- Menggabungkan semua hasil ---")
    final_ddf_optimized = dd.concat(final_ddf_list, ignore_unknown_divisions=True)

    print("\n--- Memaksa repartisi ke 1 file (Sekarang lebih aman dengan data 32-bit) ---")
    final_ddf_single_file = final_ddf_optimized.repartition(npartitions=1)

    print("Menyimpan hasil akhir ke satu file Parquet...")
    # Ganti nama file output Anda agar mencerminkan 32-bit
    output_path = '/content/drive/MyDrive/Dataset/dataset_final_100_percent_32bit.parquet'

    # Simpan Dask DataFrame yang sudah direpartisi
    final_ddf_single_file.to_parquet(
        output_path,
        engine='pyarrow',
        compression='snappy'
    )

    print(f"\n--- Selesai! ---")
    print(f"Dataset 100% 32-bit Anda ('{output_path.split('/')[-1]}') telah disimpan.")
```
